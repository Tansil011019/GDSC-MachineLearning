# Baca data dari file .csv
1. Unduh loan_sanction_train.csv dari https://www.kaggle.com/datasets/rishikeshkonapure/home-loan-approval 
2. Gunakan pandas untuk membaca file tersebut
3. Print dataframe yang telah dibaca

```python
import pandas as pd
import matplotlib.pyplot as plt
path = '/content/Dataset/loan_sanction_test.csv'
data = pd.read_csv(path)
df = pd.DataFrame(data)
display(df)
```

# Filter dan sort
1. Tampilkan data loan oleh seorang laki-laki!
2. Urutkan data berdasar jumlah pinjaman!
(cukup tampilkan hasil akhir dari dua perintah di atas)

```python
mens_data = df.loc[df.Gender == "Male"]
#ascending sort
mens_loan_sorted_asc = mens_data.sort_values(by=['LoanAmount'], ascending = True)
#descending sort
mens_loan_sorted_desc = mens_data.sort_values(by=['LoanAmount'], ascending = False)
print("Sorted Ascending.")
display(mens_loan_sorted_asc)
print("Sorted Descending.")
display(mens_loan_sorted_desc)
```

# Aggregating
1. Hitung rata-rata pendapatan dari peminjam!
2. Gender apa yang paling banyak mengajukan pinjaman, laki-laki atau perempuan?
3. Hitung rata-rata pinjaman yang diajukan untuk masing-masing area properti! 

```python
income_avg= df.ApplicantIncome.mean()
print("1. Applicant's income average {}.".format(income_avg))

print("2.", end= " ")
if(len(df[df.Gender == "Male"]) > len(df[df.Gender == "Female"])):
  print("Gender that applied for the most loans is male.")
elif(len(df[df.Gender == "Male"]) == len(df[df.Gender == "Female"])):
  print("The number of male and female applying for loans is the same.")
else:
  print("Gender that applied for the most loans is female.")

print("3. The average number of loans submitted for each property area.")
area = df.Property_Area.unique()
for property in area:
  print(property, ":", df.loc[df.Property_Area == property].LoanAmount.mean())
```
